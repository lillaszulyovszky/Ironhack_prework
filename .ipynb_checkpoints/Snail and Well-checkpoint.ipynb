```python
# import the math module  
import math 

# How long does it take to get out of the well?
def escape():
    
    # Assign problem data to variables with representative names
    # well height, daily advance, night retreat, accumulated distance
    daily_climb = [30, 21, 33, 77, 44, 45, 23, 45, 12, 34, 55]
    well_height = 125
    nightly_fall = 20
    total_distance = 0
    
    # Assign 0 to the variable that represents the solution
    total_days = 0
    
    for i in range(len(daily_climb)):

        #can get out of the well?
        if total_distance >= well_height:
            break

        #need to climb
        total_days += 1
        total_distance += daily_climb[i] - nightly_fall
        
        #print (total_distance, ' ', well_height)
      
    return total_days



# What is its maximum displacement in one day? 
def max_displacement():
    
    # Assign problem data to variables with representative names
    # well height, daily advance, night retreat, accumulated distance
    daily_climb = [30, 21, 33, 77, 44, 45, 23, 45, 12, 34, 55]
    nightly_fall = 20

    
    max_displ = 0
    for i in range(len(daily_climb)):
        displacement = daily_climb[i] - nightly_fall

        if displacement < 0: #can move down (>: move up, 0: no move)
            displacement = -displacement

        if max_displ < displacement:
            max_displ = displacement

    return max_displ

# And its minimum displacement in one day?
def min_displacement():

    # Assign problem data to variables with representative names
    # well height, daily advance, night retreat, accumulated distance
    daily_climb = [30, 21, 33, 77, 44, 45, 23, 45, 12, 34, 55]
    nightly_fall = 20

    
    min_displ = 0
    for i in range(len(daily_climb)):

        displacement = daily_climb[i] - nightly_fall        

        if displacement < 0: #can move down (>: move up, 0: no move)
            displacement = -displacement

        if i == 0 or min_displ > displacement:
            min_displ = displacement

    return min_displ

  
# What is its average speed during the day?
def avg_speed():
    
    # Assign problem data to variables with representative names
    # well height, daily advance, night retreat, accumulated distance
    daily_climb = [30, 21, 33, 77, 44, 45, 23, 45, 12, 34, 55]
    nightly_fall = 20
    
    total_sp = 0 #total speed
    for i in range(len(daily_climb)):

        total_sp += daily_climb[i] - nightly_fall

    avg_sp = 0

    if len(daily_climb) > 0:
        avg_sp = total_sp / len(daily_climb)

    return avg_sp

# What is the standard deviation of its displacement during the day?
def standard_deviation():
    
    # Assign problem data to variables with representative names
    # well height, daily advance, night retreat, accumulated distance
    daily_climb = [30, 21, 33, 77, 44, 45, 23, 45, 12, 34, 55]
    nightly_fall = 20
    
    avg_sp = avg_speed() #calculate average speed
    total_terms = 0 #total term such as daily_climb[i] - average
    
    for i in range(len(daily_climb)):
        displacement = daily_climb[i] - nightly_fall        
        total_terms += (displacement - avg_sp) * (displacement - avg_sp)

    std_dev = 0

    if len(daily_climb) > 0:
        std_dev = math.sqrt(total_terms / len(daily_climb))

    return std_dev
    
#test
print ('How long does it take to get out of the well?', escape())
print ('What is its maximum displacement in one day? ', max_displacement())
print ('What is its minimum displacement in one day? ', min_displacement())
print ('What is its average speed during the day?', avg_speed())
print ('What is the standard deviation of its displacement during the day?', standard_deviation())
```
